```python
def plotRC(ax1, dfA, dfB, crystaltype, labelA, labelB, cA, cB, xmin, xmax):
    #fig = plt.figure(figsize=(35,20))
    #ax1 = fig.add_subplot(111)
    ax1.set_xlabel("$\\omega$ ($\degree)$", fontsize=30)
    ax1.tick_params(axis='x', labelsize=30)
    ax1.set_ylabel("counts", fontsize=30)
    ax1.tick_params(axis='y', labelsize=30)
    ax1.set_title(crystaltype, fontsize=40)
    #ax1.set_yscale('log')
    #ax1.grid(True)
    ax1.plot(dfA['Omega'], dfA['counts'], label=labelA, c=cA,  linewidth=3)
    ax1.plot(dfB['Omega'], dfB['counts'], label=labelB, c=cB,  linewidth=2)
    ax1.set_xlim(xmin, xmax)
    ax1.legend(fontsize="30", frameon=False)

    #plt.show()
    return ax1
```

```python
fig = plt.figure(figsize=(35,20))
ax1 = fig.add_subplot(111)
ax = plotRC(ax1, df_SC67_A, df_SC67_B, 'MAPbBr3', 'SC67-A', 'SC67-B', 'black', 'green', 13.5, 16.3)
```

```python
fig = plt.figure(figsize=(35,20))
ax1 = fig.add_subplot(111)
ax = plotRC(ax1, df_CS104_A, df_CS106_A_220, 'CSPbBr3', 'CS104-A', 'CS106-A (220)', 'b', 'darkorange', 13.5,16)
```
