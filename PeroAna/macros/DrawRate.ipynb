```python
import matplotlib.pyplot as plt
import numpy as np

species = ("CsPbBr3 \n (Cs104-A)", "CsPbBr3 \n (Cs106)", "MAPbBr3 \n (SC67-B)")
configuration = {
    'No source': (2.08, 1.22, 0.81),
    'Am241': (5.15, 0.00, 0.00),
    'Cs137': (11.06, 0.00, 0.00),
    'Sr90': (0.00, 5.84, 2.28)
}

x = np.arange(len(species))  # the label locations
width = 0.20  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(constrained_layout=True)

for attribute, measurement in configuration.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=4)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Rate (Hz)')
ax.set_title('Event rates w/ and w/o source exposure')
ax.set_xticks(x + 1.5*width, species)
ax.legend(loc='upper left', ncols=4)
ax.set_ylim(0, 20)
``
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
species = ['No source', 'Am241', 'Cs137']
rate = [2.08, 5.15, 11.06]
bar_colors = ['tab:red', 'tab:blue', 'tab:orange']

# Plot
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(species, rate, color=bar_colors)

# Labels inside bars
for bar, value in zip(bars, rate):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() / 2,
        f'{value:.2f} Hz',
        ha='center',
        va='center',
        color='white',
        fontsize=12,
        fontweight='bold'
    )

# Axis formatting
ax.set_ylabel('Rate [Hz]', fontsize=12)
ax.set_title('Event Rates With and Without Source Exposure', fontsize=14)
ax.set_xticks([])  # Remove x-axis labels

# Optional grid
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Clean look
ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```
