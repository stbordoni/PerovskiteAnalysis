```python
import matplotlib.pyplot as plt
import numpy as np

species = ("Cs104-A", "Cs106", "SC67-B")
configuration = {
    'No source': (2.08, 1.22, 0.81),
    'Am241': (5.15, 0.00, 0.00),
    'Cs137': (11.06, 0.00, 0.00),
    'Sr90': (0.00, 5.84, 2.28)
}

x = np.arange(len(species))  # the label locations
width = 0.20  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(constrained_layout=True)

for attribute, measurement in configuration.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=4)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Rate (Hz)')
ax.set_title('Event rates w/ and w/o source exposure')
ax.set_xticks(x + 1.5*width, species)
ax.legend(loc='upper left', ncols=4)
ax.set_ylim(0, 20)

plt.show()
```
